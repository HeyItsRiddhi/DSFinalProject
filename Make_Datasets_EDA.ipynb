```python
#Make Florida Data Frame & District Data Frames

#list of voting data files
voters = ['BAK_H_20170110.txt','BAY_H_20170110.txt','BRA_H_20170110.txt','CAL_H_20170110.txt','CHA_H_20170110.txt',
          'CLA_H_20170110.txt','CLL_H_20170110.txt','CLM_H_20170110.txt','DES_H_20170110.txt','DIX_H_20170110.txt',
          'DUV_H_20170110.txt']

#list of registration data files
registration = ['BAK_20170110.txt','BAY_20170110.txt','BRA_20170110.txt','CAL_20170110.txt','CHA_20170110.txt',
                'CLA_20170110.txt','CLL_20170110.txt','CLM_20170110.txt','DES_20170110.txt','DIX_20170110.txt',
                'DUV_20170110.txt']

#Lists to store the district data frames
district_voter_hist = []
district_voter_reg = []

voter_history_df = pd.DataFrame()
for vote_file in voters: 
    file_path_v = 'Florida/Voter History/' + vote_file
    votes = pd.read_csv(file_path_v, sep="\t", header = None)
    district_voter_hist.append(votes)
    voter_history_df = voter_history_df.append(votes)
    
voter_reg_df = pd.DataFrame()
for reg_file in registration: 
    file_path_r = 'Florida/Voter Registration/' + reg_file
    reg = pd.read_csv(file_path_r, sep="\t", header = None)
    district_voter_reg.append(reg)
    voter_reg_df = voter_reg_df.append(reg)
```
